# Covid 19 Deaths in USA

The Following dataset includes a statewise distribution of deaths that have occurred in USA due to the Covid 19 pandemic.

This dataset includes death counts starting from 23rd January 2020 upto 2nd July 2021. Hence it includes most recent numbers reported by states, territories, and other jurisdictions.

*Dataset Source : https://data.cdc.gov/Case-Surveillance/United-States-COVID-19-Cases-and-Deaths-by-State-o/9mfq-cb36*

```python
# Importing pandas library
import pandas as pd

# Reading the dataset of total deaths in USA
data_deaths = pd.read_csv("/content/drive/MyDrive/Vizathon_2021_Datasets/United_States_COVID-19_Cases_and_Deaths_by_State_over_Time.csv")

data_deaths
```

```python
data_deaths.sort_values(["submission_date","state"], inplace = True) # Sorting data by subission dates and state name

data_deaths
```

```python
data_deaths.drop(["prob_cases", "new_case", "pnew_case", 
                  "prob_death", "new_death", "pnew_death", "created_at", 
                  "consent_cases", "consent_deaths"], axis = 1, inplace = True) # Dropping unwanted columns

data_deaths
```

```python
# Total cases as of 1st January 2021

import plotly.express as px

data_deaths.drop_duplicates(subset = ['state'], keep = 'first', inplace = True)
fig = px.bar(data_deaths, x = 'state', y = 'tot_cases')
fig.show()
```

```python
# Choropleth map for number of cases

import plotly.graph_objects as go

fig = go.Figure(data=go.Choropleth(
    locations=data_deaths['state'], # Spatial coordinates
    z = data_deaths['tot_cases'], # Data to be color-coded
    locationmode = 'USA-states', # set of locations match entries in `locations`
    colorscale = 'Reds',
    colorbar_title = "Cases in Million",
))

fig.update_layout(
    title_text = '2020-21 Total Cases in USA',
    geo_scope='usa', # limit map scope to USA
)

fig.show()
```

# Covid 19 Vaccinations in USA

***Introduction***

The COVID-19 vaccination campaign in the United States is an ongoing mass immunization campaign for the COVID-19 pandemic in the United States. The FDA granted emergency use authorization to the Pfizer–BioNTech vaccine on December 10, 2020; mass vaccinations began on December 14, 2020. The Moderna vaccine was granted emergency use authorization on December 17, 2020, and the Janssen (Johnson & Johnson) vaccine was granted emergency use authorization on February 27, 2021. By April 19, 2021, all U.S. states had opened vaccine eligibility to residents aged 16 and over. On May 10, 2021, the FDA approved the Pfizer vaccine for adolescents aged 12 to 15.

By July 4, 2021, 67% of the United States' adult population had received at least one dose, just short of a goal of 70%. While vaccines have helped to significantly reduce the number of daily COVID-19 infections nationwide, there are still a number of states where vaccination rates are below average, which have seen increasing numbers of cases credited to the highly-infectious Delta variant.

*Source : https://en.wikipedia.org/wiki/COVID-19_vaccination_in_the_United_States*



***Dataset***

The following dataset consists of vaccination data of USA. The dataset ranges from 12th January 2021 upto 29th July 2021

*Dataset Source : https://www.kaggle.com/paultimothymooney/usa-covid19-vaccinations*


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import plotly.express as px

# Read vaccination dataset

data_vac = pd.read_csv('/content/drive/MyDrive/Vizathon_2021_Datasets/us_state_vaccinations.csv')

data_vac.head(5)
```

```python
data_vac.drop(["total_distributed", "people_fully_vaccinated_per_hundred",
               "total_vaccinations_per_hundred", "people_fully_vaccinated",
               "people_vaccinated_per_hundred", "distributed_per_hundred",	
               "daily_vaccinations_raw", "daily_vaccinations",	"daily_vaccinations_per_million",	
               "share_doses_used"], axis = 1, inplace = True)
# Drop unwanted columns
```

```python
data_vac.drop(data_vac.loc[data_vac['location'] == "United States"].index, inplace = True) # Drop unwanted rows
```

```python
data_vac
```

```python
data_vac['month'] = pd.DatetimeIndex(data_vac['date']).month # Extracting month value

data_vac
```

```python
data_vac.drop_duplicates(['location', 'month'], keep = 'last', inplace = True) # Dropping multiple entries of the same month

data_vac
```

```python
fig  = px.scatter(data_vac, x = "total_vaccinations", y = "people_vaccinated",
                  animation_frame = "month", color = "location", 
                  hover_name = "location", log_x = True)

# Creating animated scatter plot to show vaccinations in 7 months

fig.show()
```
